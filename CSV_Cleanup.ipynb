```python
# Initial imports
import pandas as pd
```

```python
# Load in NBA data.
file_path = "Resources/nba_rankings_2014-2020.csv"
nba_df = pd.read_csv(file_path, index_col= 0)
nba_df.head(10)
```

```python
nba_df = nba_df.dropna()
```

```python
nba_19_20 = nba_df[nba_df["SEASON"] == "2019-20"]
```

```python
nba_19_20 = nba_19_20.rename(columns={"FG%":"FGP", "3PM":"TPM", "3PA":"TPA", "3P%":"TPP", "FT%":"FTP", "+/-":"PON"})
```
